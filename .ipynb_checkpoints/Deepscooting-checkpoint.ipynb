# **<u>Problem:</u> Give recommendation to a club for their next recruitment.**

The problematic we investigate is the following. 
A certain club that plays in the U.K soccer league needs to replace a player of them that was transfered to another club.
Since they were very satisfied with him, they want to replace him with a player that has, essentially, the same qualities than him at the moment of the transfer. 

The dataset we work on is the list of all players in the U.K soccer championship at the given season. However, we design our solution to satisfy more criterion:
- it has to be scalable (it has to be able to treat a much bigger database of players than the U.K championship)
- it needs to be able to answer more than this specific question (what if we want to replace another kind of player that is completely different?).

Our approach is the following:
- Basic data exploration.
- Feature engineering and data preprocessing. 
- Compress the dataset by using an autoencoder.
- Perform a clustering by using a Kmeans algorithm. The different clusters correspond to different kinds of player, we want to select the cluster that contains the player we want to replace, and then rank them accordingly to criterion given by the club.

```python
# Extract the subDataFrame corresponding to a given player position
def extractor_positions(df,role_str):
    def curry_split(string):
        return string.split(',')
    df_roles = df['roles'].apply(curry_split).apply(pd.Series)
    df_roles['id'] = df['id']
    df_roles = pd.merge(left = df,right = df_roles, how = 'inner')    #[['position',0,1,2,3,4]]
    return df_roles[(df_roles[0] == role_str) | (df_roles[1] == role_str)| (df_roles[2] == role_str)| (df_roles[3] == role_str)| (df_roles[4] == role_str)].drop(columns=['roles',0,1,2,3,4])
```

```python
print('As an example, this is the subdataframe with all the players that are M_C')
players_M_C = extractor_positions(players, 'M_C')
players_M_C
```

## Preprocessing numerical features

What I have in mind to propose a player to a club is to use a clustering algorithm. That justifies scaling all numerical features with a standard scaler, as it has been proved that empirically, it is the most appropriate method in the majority of cases. If we had more time, we could look at every feature distribution and decide more finely (min-max and robust could make sense depending on the quantity of ouliers).

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

#Only consider numerical features for this: beware, Id should not be scaled as it is not really numerical per say.
num_col = players.describe().columns
num_col_no_id = list(num_col)
num_col_no_id.remove('id')

scaler = StandardScaler()
players_scaled = pd.DataFrame(scaler.fit_transform(players[num_col].drop(columns = ['id'])),columns = num_col_no_id)
players_scaled
```

## Player recommendation algorithm.

- Extract the subdataframe corresponding to the position the club is interested in.
- Perform dimensionality reduction on the features as there are too many, even more in real-world cases, probably.
- Ask the club which player previously satisfied their needs in this position.
- Input the data of this player, then predict in which cluster the player belongs to: they are the most similar to him in the sense of the clustering.
- In this cluster, make a list of the choices that make the most sense for the club depending on their criterion.

### Naive proposition

The easiest approach to follow, both in terms of explicability and compelxity, would be the following:

- Dimensionality reduction via PCA.
- Directly apply a Kmeans.

Let us look at the covariance of the couples of features we have.

```python
import matplotlib.pyplot as plt
import seaborn as sns
correlation = players.corr()
f, ax = plt.subplots(figsize = (15, 15))
sns.heatmap(correlation, annot = True);
```

It is true that you can see some linear dependency between a few features, but it seems like the result of a reduction via PCA will not be that interesting on the long run.

### Advanced framework

- Dimensionality reduction via autoencoder (as we saw that many variables are not linearily dependent: the space of features is more likely a non-linear manifold)
- Kmeans could make sense still, but I would investigate other clustering algorithms. The biggest problem of Kmeans is that it is not robust and therefore adding new players will necessitate training from the bottom every time.

```python
# Separate the dataset: training, validation, testing.

from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(players_scaled, test_size=0.2)
X_test,X_val = train_test_split(X_test, test_size=0.5)
```

```python
players_scaled.shape
```

```python

import tensorflow as tf

from tensorflow.keras import layers, losses
from tensorflow.keras.models import Model

latent_dim = players_scaled.shape[1] #this is 34

class Autoencoder(Model):
    def __init__(self, latent_dim,n:int):
        super(Autoencoder, self).__init__()
        self.latent_dim = latent_dim   
        self.encoder = tf.keras.Sequential([
          layers.Dense(latent_dim, activation='relu'),
          layers.Dense(26, activation='relu'),
          layers.Dense(n, activation='relu'),
          layers.Dense(6, activation='tanh')
            
        ])
        self.decoder = tf.keras.Sequential([
            layers.Dense(n, activation='relu'),
            layers.Dense(26, activation='relu'),
          layers.Dense(latent_dim, activation='relu')
        ])

    def call(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

autoencoders_list = [Autoencoder(latent_dim,n) for n in range(6,26)]

for autoencoder in autoencoders_list:
    autoencoder.compile(optimizer='adam', loss=losses.MeanSquaredError())
#We then compare all of these in terms of performance.
```

```python
#After investigation, n=19 provides the best result.
autoencoder = Autoencoder(latent_dim,n=19) 
autoencoder.compile(optimizer='adam', loss=losses.MeanSquaredError())

hist_auto = autoencoder.fit(X_train,X_train,batch_size=16,epochs=100, shuffle=True, validation_data=(X_val,X_val))

# Summarize history for loss
plt.figure()
plt.plot(hist_auto.history['loss'])
plt.plot(hist_auto.history['val_loss'])
plt.title('Autoencoder model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
loss_list = []
for autoencoder in autoencoders_list:
    hist_auto = autoencoder.fit(X_train,X_train,batch_size=16,epochs=70, shuffle=True, validation_data=(X_val,X_val))
    loss_list.append((hist_auto.history['loss'],hist_auto.history['val_loss']))
```

On this autoencoder that has not even been optimized, I already have a decent gain in compression of features, and I did not have to push the network really deep (not that many features to work with anyways, so not necessary ).

```python
# Evaluate the model on the test data using `evaluate`
print("Evaluate on test data")
results = autoencoder.evaluate(X_test,X_test, batch_size=32)
print("test loss, test acc:", results)
```

```python
autoencoder.summary()
```

Let us look at the encoded dataframe.

```python
players_encoded = pd.DataFrame(autoencoder.encoder(np.array(players_scaled)).numpy(),columns = [f'latent_feature{i}' for i in range(0,6) ])
players_encoded#['latent_feature12'].unique()
```

Let us perform as simple KNN to conclude the study.

```python
score_auto = []
range_values = range(1, 20)
for i in range_values:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(players_encoded)
    score_auto.append(kmeans.inertia_)
    if i %5 == 0:
        print(f'{i} steps completed')
```

```python
plt.plot(score_auto, 'bx-')
plt.title('Find the right number of cluster')
plt.xlabel('Clusters')
plt.ylabel('Scores WCSS')
plt.show()
```

For this example, the elbow method leads to think that we can go with 5 clusters.

```python
#--k-means++ ensures not falling into random initialization trap
kmeans_auto = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
labels_auto = kmeans_auto.fit_predict(players_scaled)
```

```python
players_cluster_encoded = pd.concat([players_encoded, pd.DataFrame({'cluster': labels_auto})], axis = 1)
players_cluster_encoded
```
